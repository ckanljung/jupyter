## Temperature curve for 2018 Christmas Ham

125 Degrees Celsius oven temperature

- 2018 with foil
- 2019 without foil


```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline
```

```python
def baseline(vals):
    offseter = lambda t: t - vals[1]
    return np.array([offseter(xi) for xi in vals])
```

```python

xval18 = np.array([6, 15, 20, 26, 41, 50, 53, 56, 
                 61, 64, 60+21, 60+48, 60+51, 60+53, 60+59, 120+9,
                 120+12, 120+14, 120+16, 120+42,
                 180, 180+2, 180+6, 180+10, 180+24, 180+29, 180+31, 180+35, 180+40, 180+44, 180+48, 180+52,
                 240+1, 240+3])
yval18 = np.array([8, 8, 9, 9, 12, 14, 15, 16, 
                 17, 18, 23, 33, 34, 35, 37,
                 40, 41, 41, 42, 50,
                 55, 56, 57, 57, 61, 62, 62, 63, 65, 65, 66, 67,
                 68, 69])
xval_off18 = baseline(xval18)

xval19 = np.array([17,26,35,38,41,44,51,54,56,60+1, 60+7, 60+13, 60+14, 60+15, 60+18, 60+24, 60+29, 60+30,
                  60+35, 60+37, 60+38, 60+41, 60+48, 60+53, 60+59, 120+4, 120+6, 120+14, 120+18, 120+28,
                  120+30, 120+33, 120+44, 120+46, 120+52, 120+58, 120+59, 180+3, 180+6, 180+8, 180+15, 
                  180+17, 180+22])
yval19 = np.array([9,10,11,12,12,14,16,17,18,20,22,26,27,27,28,31,33,34,36,37,38,39,42,44,47,49,50,
                  53,55,58,58,59,62,63,64,66,67,68,68,69,70,71,72])

xval_off19 = baseline(xval19)
```

```python

plt.rcParams['figure.figsize'] = (10, 10)
plt.rcParams['font.size'] = '14'
plt.title('Ham temperature curve')
plt.xlabel('Time in oven')
plt.ylabel('Temperature')
plt18, = plt.plot(xval_off18, yval18)
plt19, = plt.plot(xval_off19, yval19)
plt.legend([plt18,plt19], ['2018','2019'])
plt.show()
```
